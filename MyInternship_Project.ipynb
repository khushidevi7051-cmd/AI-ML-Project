##**Project:**  Pneumonia Detection via Chest X-Ray Classification



##***Project Overview***
 This project establishes an end-to-end Machine Learning pipeline to classify chest X-ray images into two binary categories: **Normal (0)** and **Pneumonia (1)**.

 The Pipeline handles:

*  **Image Ingestion & Preprocessing**
(grayscale conversion,resizing,flattening)
*   **Dataset Partitioning**
(Stratified train/test split)
*   **Deep Learning Model Construction**
(Multi-Layer Perceptron using Keras)
*   **Itervative Model Training & Performance Evaluation**





#**Step-by-Step Implementation**

##Step 1: Import Required Libraries
You need to import core utilities for file handling, numerical operations, image processing, and deep learning.

```python
import os
import numpy as np
from PIL import Image
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

##Step 2: Download Dataset From GitHub

```python
!git clone https://github.com/YBIFoundation/ComputerVision.git
```

##Step 3: Define Dataset Directiories
You set the file paths pointing to where your source images are saved on your computer or cloud environment.

```python
normal_dir = '/content/ComputerVision/NormalXray'
pneumonia_dir='/content/ComputerVision/PneumoniaXray'
```

##Step 4: Initialize Variables and Constants
You define the target size for the images and create empty lists to hold your processed data.

```python
IMG_SIZE=(64,64)
X=[]
y=[]
```

##Step 5: Process the "Normal" Images
You loop through every image file inside the normal directory to read,clean and store it.

```python
for file in os.listdir(normal_dir):
    img = Image.open(os.path.join(normal_dir, file)).convert("L")
    img= img.resize(IMG_SIZE)
    X.append(np.array(img).flatten())
    y.append(0)
```

##Step 6: Process the "Pneumonia" Images
You repeat the exact same processing loop for the diseased X-Ray images,but assign them a different label.

```python
for file in os.listdir(pneumonia_dir):
    img = Image.open(os.path.join(pneumonia_dir, file)).convert("L")
    img= img.resize(IMG_SIZE)
    X.append(np.array(img).flatten())
    y.append(1)
```

##Step 7: Convert Lists to Numpy Arrays
Python lists are slow for machine learing, so you tranform them into highly optimized NumPy arrays.

```python
X=np.array(X)
y=np.array(y)
```

##Step 8: Verify Data Shapes
You print out the final structure of your dataset to ensure everything loaded correctly before building a model.

```python
print(X.shape)
print(y.shape)
```

##Step 9: Data Inspection and Feature Verification

```python
print("Total flattened pixels:",64*64)
print("\nFirst image pixels values\n:",X)
print("\nAll target labels\n:", y)

```

##Step 10: Split Data into Training and Testing Sets
You divide your dataset into two parts: one to train the model,and one to test how well it learned.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2529, stratify=y)
```

##Step 11: Initialize the Sequential Model
You create a linear backbone container where neural network layers can be stacked on top of each other.

```python
model = Sequential()
```

##Step 12: Add Hidden Layers
You add fully connected(Dense) layers that will look for patterns in the flattened image pixel data.

```python
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
```

##Step 13: Add the Output Layer
You add final layer with single neuron to give the final classification answer.

```python
model.add(Dense(1,activation='sigmoid'))
```

##Step 14: Compile the Model
Before training you configure how the model learns by setting its mathematical goals and evaluation metrics.

```python
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

##Step 15:Train (Fit) the Model
You feed the training data into the network so it can learn by trial and error.

```python
model.fit(X_train, y_train, epochs=10)
```

##Step 16: Review Model Structure & Predict
You check the final blueprint layout of your network and generate predictions on the unseen test images.

```python
model.summary()
y_pred=np.round(model.predict(X_test))
```

##Step 17: Generate a Classification Report
You import and print a detailed breakdown of how well your model performed on the test data.

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
